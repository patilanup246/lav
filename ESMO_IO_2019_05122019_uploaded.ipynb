```python
def getProperty(info):
    data = info.parent.find('div',attrs={'class':'property'}).get_text().strip()
    return data
   
def getModerators(info):
    moderators = []
    affilation = []
    texts = info.parent.find('div',attrs={'class':'property'}).get_text().strip().split("\n")
#     print(parent)
    for line in texts:
        words = line.split("(")
#         print(lis)
        moderators.append(words[0].strip())
        affilation.append(words[1][:-1].strip())
#         print("Author" + lis[0].strip())
#         print("Affilation" + lis[1][:-1].strip())
    return ";".join(moderators), ";".join(affilation)
    
def getStartEndTime(info):
    data = info.parent.find('div',attrs={'class':'property'}).get_text().strip().split("-")
    return data[0].strip(), data[1].strip()

import re
def format_html(ip):
    ip = str(ip)
    ip = re.sub("(<table>).*?(</table>)", '', ip)
    ip = re.sub("(<table-wrap-foot>).*?(</table-wrap-foot>)", '', ip)
    return ' '.join(BeautifulSoup(ip).text.split(' '))
```
